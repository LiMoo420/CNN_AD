```python

val_acc = [0.5415, 0.5603, 0.5915, 0.6870, 0.7199, 0.7684, 0.8388, 0.8826, 0.9155, 0.9405, 0.9624, 0.9186, 0.9624, 0.9202, 0.9734]

val_loss = [0.4211, 0.3327, 0.3934, 0.2323, 0.1175, 0.3011, 0.1112, 0.1728, 0.1271, 0.0688, 0.0155, 0.1244, 0.0053, 0.0998, 0.0032]

epochs_range = range(15)

plt.figure(figsize=(8,8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title("Training and Validation Accuracy")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
